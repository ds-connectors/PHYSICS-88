```python
# This is based on Avi's work
# http://datahub.berkeley.edu/user-redirect/interact?account=avirukt&repo=dipole_demo&branch=master&path=dipole.ipynb
# There was one exponent that needed to be changed, and I deleted some material.

from mpl_toolkits.mplot3d import Axes3D #import the class for 3d plotting
from matplotlib import pyplot as plt
import numpy as np
```

```python
%matplotlib notebook
#set the plots to display interactively (useful for 3d plotting, you can rotate the view by dragging)

N=8
x = np.linspace(-2,2,N)
x, y, z = np.meshgrid(x,x,x) #create 3d grid for x,y,z values
rplus = np.sqrt(x**2 + y**2 + (z-1)**2)
rminus = np.sqrt(x**2 + y**2 + (z+1)**2)
E = np.array([x,y,z-1])/rplus**3.0 - np.array([x,y,z+1])/rminus**3.0
for i in range(N**3):
    i1 = i // N**2
    i2 = (i//N) % N
    i3 = i % N
    Eabs = np.sum(E[:,i1,i2,i3]**2)**.5
    if Eabs > 4/N:
        E[:,i1,i2,i3] /= Eabs*N/4
fig = plt.figure(figsize=(4,4)) #create figure object on which the plot will be renderred
ax = Axes3D(fig) #create Axes3D object to display the plot
ax.quiver(x, y, z, *E, color="k", alpha = 0.5)
```
