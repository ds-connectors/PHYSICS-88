## Interactive 3D Plots

```python
col1,col2,col3 = np.loadtxt('PossiblyDifferentHousingData_3columns.txt', unpack = True)
len(col3)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook
#set the plots to display interactively (useful for 3d plotting,
#you can rotate the view by dragging)

fig = plt.figure(figsize=(4,4))
ax = Axes3D(fig)
ax.scatter(col1,col2,col3, c='r', marker='o')
ax.set_xlabel('Bus')
ax.set_ylabel('Grocery')
ax.set_zlabel('Rent')
```
